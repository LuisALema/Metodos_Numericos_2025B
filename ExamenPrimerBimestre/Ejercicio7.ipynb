```python
def Spline(x: float, x0: float, pars: dict[str, float]) -> float:
    A = pars["a"]
    B = pars["b"]
    C = pars["c"]
    D = pars["d"]
    dx = x - x0  
    return A + B * dx + C * dx**2 + D * dx**3
```

**Grafica de la solucion**

```python
import matplotlib.pyplot as plt
import numpy as np

# Puntos
xs = [-1, 0, 1]
ys = [1, 5, 3]

# Splines con pendiente m = -3 en x1 = 0
coef = [
    {"a": 1, "b": 12, "c": -7, "d": -1},   # S0 en [-1,0]
    {"a": 5, "b": -3, "c": -4, "d": 3},   # S1 en [0,1]
]

# Graficar las dos curvas spline
for idx, xi in enumerate(xs[:-1]):
    x_segment = np.linspace(xi, xs[idx + 1], 80)
    y_segment = Spline(x_segment, xi, coef[idx])
    plt.plot(x_segment, y_segment, color="red")

plt.scatter(xs, ys, color="blue")

m = -3
x_line = np.linspace(-1, 1, 50)
y_line = 5 + m*(x_line - 0)
plt.plot(x_line, y_line, "b--")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolación con Splines Cúbicos (m = -3)")
plt.grid(True)
plt.show()
```
