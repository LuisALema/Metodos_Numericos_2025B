**Autor:** Luis Lema  
**Modelo utilizado:** ChatGPT – GPT-5.1  
**Modo de uso:** Interfaz web (ChatGPT en navegador)




##  Introducción

El objetivo de esta actividad es identificar casos en los que un **modelo de lenguaje de gran tamaño (LLM)** produce respuestas incorrectas.  

Para ello se prepararon **tres prompts**, cada uno con una única respuesta correcta, lo cual permite detectar claramente los errores cometidos por el modelo.  
En cada ejemplo se incluye:

- Prompt utilizado  
- Respuesta incorrecta del LLM  
- Explicación del error  
- Espacio para captura de pantalla
